**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Data**


```python
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")
train.head()
```

**Explore Target Distribution**

```python
sns.countplot(data=train, x="target")
plt.title("Distribution of Target Labels")
plt.show()

print(train['target'].value_counts())
```

**Check for Nulls**

```python
train.isnull().sum()
```

**Check Text Lengths**

```python
train['text_len'] = train['text'].apply(len)
sns.histplot(train['text_len'], bins=30)
plt.title("Text Length Distribution")
plt.show()
```

**Word Cloud**

```python
from wordcloud import WordCloud

disaster_texts = " ".join(train[train['target'] == 1]['text'])
nondisaster_texts = " ".join(train[train['target'] == 0]['text'])

WordCloud(width=800, height=400).generate(disaster_texts).to_image()

```

```python
WordCloud(width=800, height=400).generate(nondisaster_texts).to_image()
```
